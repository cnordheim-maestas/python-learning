# Pandas DataFrames

```python
import pandas as pd

data = pd.read_csv("gapminder_gdp_europe.csv", index_col="country")
print(data.iloc[0,0])
```

```python
# get location by using .loc with labels instead of numerical position

print(data.loc["Albania", "gdpPercap_1952"])
```

```python
# use : to get all of a column or row value

print(data.loc["Albania", :])
```

```python
print(data.loc[:, "gdpPercap_1952"] )
```

```python
# Get multiple columns and multiple rows

print(data.loc['Italy':'Poland', 'gdpPercap_1962':'gdpPercap_1972']) #row:row, column:column
```

```python
print(data.loc['Italy':'Poland', 'gdpPercap_1962':'gdpPercap_1972'].max()) #row:row, column:column
```

```python
# Use a subset of data to keep output readable

subset = data.loc['Italy':'Poland', 'gdpPercap_1962':'gdpPercap_1972']
print('subset:\n', subset) # \n to add a line break
print('\nWhere are values larger?\n', subset > 10000)
```

```python
mask = subset > 10000
print(subset[mask])
```

```python
data = pd.read_csv('gapminder_gdp_oceania.csv', index_col='country') 
#Let's make the "years" less long and replace the gpdpercap nonsense!
years=data.columns.str.replace('gdpPercap_','')

#convert to integers from strings
years=years.astype(int)

years

#now let's make this prettier string in our data
data.columns = years
```

```python
data.loc['Australia'].plot()
```

```python
data.plot()
```

```python
data.T.plot() #T to transpose
```

```python
data.T.plot() #pandas
plt.ylabel('gdp') #matplotlib
# fun fact: these commands are from different libraries
# if you create with pandas you can use matplotlib to adjust it
```

```python
# Barplot
data.T.plot(kind='bar')
plt.ylabel('gdp')
plt.style.use('ggplot')
```

```python
gdp_australia = data.loc['Australia']
years = data.columns

plt.plot(years, gdp_australia, 'g--') #(x,y, green and dashed)
```

```python
gdp_australia = data.loc['Australia']
gdp_nz = data.loc['New Zealand']
years = data.columns

plt.plot(years, gdp_australia, 'g--') #(x,y, green and dashed)
plt.plot(years, gdp_nz, 'b-') #(x,y, blue and not dashed)
```

```python
# Add a legend
gdp_australia = data.loc['Australia']
gdp_nz = data.loc['New Zealand']
years = data.columns

plt.plot(years, gdp_australia, 'g--', label= "Australia") #(x,y, green and dashed)
plt.plot(years, gdp_nz, 'b-', label="NZ") #(x,y, blue and not dashed)
plt.legend(loc='lower right')
```

```python
# Let's save it!

gdp_australia = data.loc['Australia']
gdp_nz = data.loc['New Zealand']
years = data.columns

plt.plot(years, gdp_australia, 'g--', label= "Australia") #(x,y, green and dashed)
plt.plot(years, gdp_nz, 'b-', label="NZ") #(x,y, blue and not dashed)
plt.legend(loc='lower right')

plt.savefig('oceania.png') #save it to current directory
```

```python
def print_date(year, month, day):
    joined = str(year) + '/' + str(month) + '/' + str(day)
    print(joined)
    
print_date(1871, 3, 19) #year, month, day
```

```python
def average(values):
    if len(values)==0:
        return None
    return sum(values)/len(values)
```

```python
a= average([1,3,4])
print('the average is: ', a)
```
